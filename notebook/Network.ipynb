```julia
"""
expectation:`e`、amplitude:`a`、wavelength:`λ`、time steps:`t_max`の2準位の波`Vector{Int}`を返す。
"""
function two_level_N_vec(e::Int, a::Int, λ::Int, t_max::Int)::Vector{Int}
    e .+ a .* [((x - 1) % λ) < (λ / 2) ?  -1 : 1 for x in 1:t_max]
end

two_level_N_vec(400, 100, 6, 100)
```

```julia
# ToDo: 平均が1になるようにする。
function make_simple_growth_rate_vec(quarter_cycle_length::Int, growth_rate_diff::Float64, t_max::Int)::Vector{Float64}
    growth_rate_vec = fill(0.0, t_max)

    rise_flag = false  # true ならばgrowth_rate_diff分growth_rateが増加、falseならばgrowth_rate_diff分growth_rateが減少

    for t in 1:t_max
        # half_cycle_length の最初でriseフラグを反転
        t % half_cycle_length == 1 && (rise_flag = !rise_flag)

        growth_rate_vec[t] = rise_flag ? 1 + growth_rate_diff : 1 / (1 + growth_rate_diff)
    end

    growth_rate_vec
end

@show growth_rate_vec = make_simple_growth_rate_vec(3, 0.2, 13)
@show growth_rate_vec = make_simple_growth_rate_vec(3, 0.2, 12)
@show growth_rate_vec = make_simple_growth_rate_vec(10, 0.1, 567)

population = 100
for growth_rate in growth_rate_vec
    @show population *= growth_rate
end
```
