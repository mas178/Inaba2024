# Cooperation Heatmap (Payoff Table)

```julia
using DataFrames: DataFrame, AbstractDataFrame, groupby
using Plots: Plot, plot, heatmap, savefig

include("../src/SimPlot.jl")
using .SimPlot
```

```julia
function plot_heatmap(df::AbstractDataFrame, symbol::Symbol, title::String)::Plot
    β = sort(unique(df.β))
    σ = sort(unique(df.σ))

    return heatmap(
        σ,
        β,
        reshape([SimPlot.get_value(df, s, b, :σ, :β, symbol) for b in β, s in σ], length(β), length(σ)),
        xlabel = "σ",
        ylabel = "β",
        xlims = (-0.05, 1.05),
        ylims = (-0.05, 1.05),
        xticks = 0:0.1:1,
        yticks = 0:0.1:1,
        # clims = (0.0, 1.0),
        title = title,
    )
end;
```

```julia
# df_vec = SimPlot.csv_to_df(["HBESJ4_1_CooperationHeatmapPayoff"]);
df_vec = SimPlot.csv_to_df(["20231128_211537", "20231129_064710", "20231129_134014"]);
```

```julia
for df in df_vec
    df.initial_T .= df.T[1]
end;
```

```julia
columns = names(df_vec[1])
key_columns = columns[[1; 52; 3:11]]
value_columns = columns[[2; 14:51]]
mean_df = SimPlot.make_mean_df(df_vec, key_columns, value_columns)
empty!(df_vec)
mean_df
```

```julia
plot(
    plot_heatmap(mean_df, :cooperation_rate, "cooperation_rate"),
    plot_heatmap(mean_df, :T, "T"),
    plot_heatmap(mean_df, :weak_C1, "weak_C1"),
    plot_heatmap(mean_df, :weak_component1_count, "weak_component1_count"),
    plot_heatmap(mean_df, :medium_component1_count, "medium_component1_count"),
    plot_heatmap(mean_df, :strong_component1_count, "strong_component1_count"),
    plot_heatmap(mean_df, :weak_component1_size_μ, "weak_component1_size_μ"),
    plot_heatmap(mean_df, :medium_component1_size_μ, "medium_component1_size_μ"),
    plot_heatmap(mean_df, :strong_component1_size_μ, "strong_component1_size_μ"),
    plot_heatmap(mean_df, :weak_component1_size_max, "weak_component1_size_max"),
    plot_heatmap(mean_df, :medium_component1_size_max, "medium_component1_size_max"),
    plot_heatmap(mean_df, :strong_component1_size_max, "strong_component1_size_max"),
    layout = (4, 3),
    size = (1100, 1100),
)
```

```julia
savefig(plot_heatmap(mean_df, :cooperation_rate, ""), "../img/HBESJ/4_1_CooperationHeatmapPayoff_cooperation_rate.png")
savefig(plot_heatmap(mean_df, :strong_component1_count, ""), "../img/HBESJ/4_2_CooperationHeatmapPayoff_strong_component1_count.png");
```
